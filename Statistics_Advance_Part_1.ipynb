```python
#statistics part 1
Q1: What is a random variable in probability theory?
A1: A random variable is a variable that takes on different values based on the outcomes of a random experiment. It maps outcomes of a probability experiment to numerical values.

Q2: What are the types of random variables?
A2: There are two main types:

Discrete Random Variable: Takes countable values (e.g., number of heads in coin tosses).

Continuous Random Variable: Takes infinite, uncountable values within a range (e.g., height, weight).

Q3: What is the difference between discrete and continuous distributions?
A3:

Discrete Distribution: Deals with countable outcomes.

Continuous Distribution: Deals with outcomes that can take any value in an interval.

Q4: What are probability distribution functions (PDF)?
A4: A PDF gives the probability that a random variable takes on a specific value (for discrete variables) or lies within a certain range (for continuous variables).

Q5: How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
A5:

PDF: Shows the probability of a specific value (or density for continuous).

CDF: Shows the probability that a random variable is less than or equal to a value.

Q6: What is a discrete uniform distribution?
A6: It is a distribution where all outcomes are equally likely. For example, each face of a fair die has a 1/6 probability.

Q7: What are the key properties of a Bernoulli distribution?
A7:

Only two outcomes: success (1) or failure (0)

One trial

Probability of success = p, failure = 1 - p

Q8: What is the binomial distribution, and how is it used in probability?
A8:
It models the number of successes in a fixed number of independent Bernoulli trials. It's used when each trial has the same probability of success.

Q9: What is the Poisson distribution and where is it applied?
A9:
It models the number of times an event occurs in a fixed interval of time or space. Used for rare events, like customer arrivals per hour or phone calls received.

Q10: What is a continuous uniform distribution?
A10:
A distribution where every value in a continuous range is equally likely. Example: Picking a random number between 0 and 1.

Q11: What are the characteristics of a normal distribution?
A11:

Bell-shaped curve

Symmetrical around the mean

Mean = Median = Mode

Follows empirical rule (68%-95%-99.7%)

Q12: What is the standard normal distribution, and why is it important?
A12:
A normal distribution with a mean of 0 and a standard deviation of 1. It is important because it simplifies calculations and is used to compute Z-scores.

Q13: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
A13:
The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution. It's critical because it allows for normal-based inference.

Q14: How does the Central Limit Theorem relate to the normal distribution?
A14:
The CLT explains why the normal distribution is so commonly used — it applies to sample means even when the original data is not normally distributed.

Q15: What is the application of Z statistics in hypothesis testing?
A15:
Z statistics help determine whether to reject the null hypothesis by comparing the Z-score of the sample to a critical value.

Q16: How do you calculate a Z-score, and what does it represent?
A16:
Z = (X - μ) / σ
It represents how many standard deviations a value is from the mean.

Q17: What are point estimates and interval estimates in statistics?
A17:

Point Estimate: A single value estimate of a parameter (e.g., sample mean).

Interval Estimate: A range of values (confidence interval) likely to contain the parameter.

Q18: What is the significance of confidence intervals in statistical analysis?
A18:
Confidence intervals provide a range that likely contains the true population parameter, giving a measure of reliability for estimates.

Q19: What is the relationship between a Z-score and a confidence interval?
A19:
Z-scores are used to calculate confidence intervals for normally distributed data. The interval depends on the Z value for the desired confidence level.

Q20: How are Z-scores used to compare different distributions?
A20:
Z-scores standardize values, allowing comparison across different distributions by showing how far a value is from its own mean in standard deviation units.

Q21: What are the assumptions for applying the Central Limit Theorem?
A21:

Independent random samples

Sample size is sufficiently large (typically n ≥ 30)

Finite population variance

Q22: What is the concept of expected value in a probability distribution?
A22:
The expected value is the long-term average outcome of a random variable. It is calculated by summing all possible values weighted by their probabilities.

Q23: How does a probability distribution relate to the expected outcome of a random variable?
A23:
The probability distribution shows the likelihood of each outcome. The expected outcome is the weighted average (expected value) based on this distribution.

```
